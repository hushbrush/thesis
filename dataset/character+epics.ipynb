```python
def find_second_occurrence(text, pattern, occurence_no):
    matches = [m.start() for m in re.finditer(re.escape(pattern), text)]
    return matches[occurence_no] if len(matches) > 1 else None
```

```python
# Get slicing points for each work
start_works = find_second_occurrence(hesiod_text, "HESIOD’S WORKS AND DAYS", 1)
end_works = find_second_occurrence(hesiod_text, "THE DIVINATION BY BIRDS", 1)

start_theogony = find_second_occurrence(hesiod_text, "THE THEOGONY", 1)
end_theogony = find_second_occurrence(hesiod_text, "THE CATALOGUES OF WOMEN AND EOIAE", 1)

start_shield = find_second_occurrence(hesiod_text, "THE SHIELD OF HERACLES", 1)
# Assuming end of text is the end of Shield (or until "THE MARRIAGE OF CEYX" if you prefer)
end_shield = hesiod_text.find("THE MARRIAGE OF CEYX")  # Optional: change as needed

# Now extract each section
works_text = hesiod_text[start_works:end_works].strip()
theogony_text = hesiod_text[start_theogony:end_theogony].strip()
shield_text = hesiod_text[start_shield:end_shield].strip()

# Confirm lengths or a preview
print(f"Works length: {len(works_text)}")
print(f"Theogony length: {len(theogony_text)}")
print(f"Shield length: {len(shield_text)}")
```

```python
books['source_title'] = books['source_title'].apply(
    lambda x: next((ft for ft in female_characters['source_title'] if x.strip().lower() == ft.strip().lower()), x)
)
```

```python
from rapidfuzz import process, fuzz

def get_best_match(title, choices, threshold=90):
    match = process.extractOne(title, choices, scorer=fuzz.ratio)
    if match and match[1] >= threshold:
        return match[0]
    return title

# Get the list of clean titles from female_characters
female_titles = female_characters['source_title'].unique()

# Apply fuzzy matching to each title in epic_books
books['source_title'] = books['source_title'].apply(
    lambda x: get_best_match(x, female_titles)
)
```

```python

def normalize_title(title):
    title = title.lower()
    title = re.sub(r'^(the|a)\s+', '', title)  # remove starting 'the' or 'a'
    title = re.sub(r'^(epic of|song of)\s+', '', title)  # remove leading phrases
    title = re.sub(r'[^\w\s]', '', title)  # remove punctuation
    title = title.strip()
    return title

# Apply normalization to both DataFrames
books['normalized_title'] = books['source_title'].apply(normalize_title)
female_characters['normalized_title'] = female_characters['source_title'].apply(normalize_title)

# Optionally print for verification
print(books[['source_title', 'normalized_title']].sort_values('normalized_title').reset_index(drop=True))
print(female_characters[['source_title', 'normalized_title']].sort_values('normalized_title').reset_index(drop=True))
```
